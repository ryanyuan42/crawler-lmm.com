```python
def get_lineMsg(s, page):
	url = 'http://www.lvmama.com/youlun/line-l660-s0-b18702-m0.html?' + 'p=' + str(page)
	content = s.get(url).content.decode('utf-8')
	cruise_msg = re.findall(re.compile(r'(<h4>)(.*?>)(.*?)<'), content)

	date = re.findall(re.compile(r'开航日期：(\d/\d+)'),content)
	cruise = re.findall(re.compile(r'【(.*?)】'), content)
	hotel = []
	price = []
	link = []
	soup = BeautifulSoup(content, "html.parser")

	for header in soup.find_all("h4"):
		try:
			link.append(re.findall(re.compile(r'(href=")(.*?)("\s)'), str(header))[0][1])
		except IndexError as e:
			print("cannot find the link")

	print('访问网络中...')
	item = 1
	for l in link:
		hotel_type = s.get(l).content.decode('utf-8')
		hotel.append(re.findall(re.compile(r'data-branchName="(.*)"\r'), hotel_type))
		price.append(re.findall(re.compile(r'data-fstPrice="(.*)"\r'), hotel_type))
		print("item {0} scaned".format(item))
		item += 1


	return date, cruise, hotel, price
```

```python
def run():
	s = requests.session()	
	filename = input('请输入你想保存数据的文件名（example: xxx.csv）: ')
	all_page = int(input('请输入你需要抓取的页数 : '))
	page = 1 	
	while page <= all_page:
		dates, cruises, hotels, prices = get_lineMsg(s, page)
		file = open(filename, 'w')
		for date, cruise, hotel, price in zip(dates, cruises, hotels, prices):
			if hotel and price:
				for h, p in zip(hotel,price):
					file.write(date + ',' + cruise + ',' + h + ',' + p) 
					file.write("\n")
		print('第 %s 页爬虫结束...' % (page))
		page += 1

	file.close()
```
